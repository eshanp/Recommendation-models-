```python
moviemat = df.pivot_table(index='Customer_ID',columns='Movie Name',values='Rating')
moviemat.head()
```

```python
moviemat
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
TheGodfather_user_ratings = moviemat['The Godfather']
ForrestGump_user_ratings = moviemat['Forrest Gump']
```

```python
TheGodfather_user_ratings
```

```python
similar_to_thegodfather = moviemat.corrwith(TheGodfather_user_ratings)
similar_to_forrestgump = moviemat.corrwith(ForrestGump_user_ratings)
```

```python
corr_thegodfather = pd.DataFrame(similar_to_thegodfather,columns=['Correlation'])
corr_thegodfather.dropna(inplace=True)
corr_thegodfather.head()
```

```python
corr_thegodfather.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_thegodfather = corr_thegodfather.join(ratings['num of ratings'])
corr_thegodfather.head()
```

```python
corr_thegodfather[corr_thegodfather['num of ratings']>150].sort_values('Correlation',ascending=False)
```

```python
corr_thegodfather
```

```python
corr_forrestgump = pd.DataFrame(similar_to_forrestgump,columns=['Correlation'])
corr_forrestgump.dropna(inplace=True)
corr_forrestgump = corr_forrestgump.join(ratings['num of ratings'])
corr_forrestgump[corr_forrestgump['num of ratings']>150].sort_values('Correlation',ascending=False)
```
